# Correlation
* What is Correlation
* Variable wihtin a dataset can be related for lots of reasons
* Types:
 - Pearson
 - Spearman's rho
 - Kendall's tau

## For Example
1. One variable could cause or  depend on the value of another variable.
2. One variable could be lightly associated with another variable.
3. Two variables could depend on a third unknown variable.

**Positive Correlation**: both variables chnage in the same direction.

**Nuetral Coorelation**: No relationship in the change of the variables.

**Negative Coorelation**: variables change in opposite directions.

# Coavriance
* Variables can be realted by a linear relationship. This is a relationship that is consistently additive across the two data samples.
* The realtionship can be summarized between two variables, called th covariance.
* The sign of the covariance can be interperated as whether the two variables change in the same dirction (positive) or change in different directions (negative)
* The magnitude of the is not easily interperated. A covariance value of zero indicates that both variables are completely independent.

```python
# Python code to demonstarte the 
# use of numpy

import numpy as np

x = [1.23, 2.12, 3.34, 4.5]
y = [2.56, 2.89, 3.76, 3.95]

# find out covariance with respect to columns
cov_mat = np.stack((x, y), axis = 0)
cov_mat 
print(np.cov(cov_mat))
```
